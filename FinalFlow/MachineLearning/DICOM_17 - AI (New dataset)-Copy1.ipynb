# Data Visualization

```python
def performance_metrics(y_test, y_pred):
    print('Performance metrics:')
    print(classification_report(y_test, y_pred), '\n')
    print('Confusion Matrix:')
    print(confusion_matrix(y_test, y_pred),'\n')
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    print('True positive: ', tp)
    print('True negavtive: ', tn)
    print('False positive: ', fp)
    print('False negative: ', fn)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.3, random_state=0, shuffle = False)
```

```python
df_testing = pd.read_csv('/Users/rt/Desktop/College/Cancer Project/data/dataset_testing_trial.csv')
output_testing = df_testing['Status']
df_testing.drop(columns = ['Patient', 'Slice no', 'Distance', 'Status', 'Unnamed: 0'], inplace = True)
cols = df_testing.columns
scalar = StandardScaler()
temp = scalar.fit_transform(df_testing)
df_testing = pd.DataFrame(temp, columns = cols)
display(df_testing)
```

```python
def sample_case_performance(classifier):
    predicted_values = classifier.predict_classes(df_testing).tolist()
    predicted_values = [i[0] for i in predicted_values]
    _predicted = []
    _target = []
    print()
    for i, j in zip(range(0, len(predicted_values), 3), range(0, len(output_testing), 3)):
        _predicted.append(mode(predicted_values[i:i+3]))
        _target.append(mode(output_testing[i:i+3].tolist()))
    correct = 0
    incorrect = 0
    count = 0
    for i, j in zip(_predicted, _target):
        count += 1
        if i == j:
            correct += 1
            print(f'CORRECT SLICE NUMBER : {count} | PREDICTED OUTPUT : {i}')
        else:
            incorrect += 1
            print(f'INCORRECT SLICE NUMBER : {count} | PREDICTED OUTPUT : {i}')
    accuracy = correct/(correct + incorrect)
    print()
    print(f'Correct predictions : {correct}')
    print(f'Incorrect predictions : {incorrect}')
    print(f'Accuracy : {accuracy*100}')
```

# Neural network

## Model - 1

```python
def model_1(_batch_size, _kernel_constraint, loss_function):
    classifier1 = Sequential([
        Dense(7, activation = 'relu', kernel_initializer = 'random_normal', input_dim = 7, kernel_constraint = _kernel_constraint),
        Dense(6, activation = 'relu', kernel_initializer = 'random_normal', kernel_constraint = _kernel_constraint),
        Dense(1, activation = 'sigmoid', kernel_initializer = 'random_normal', kernel_constraint = _kernel_constraint)
    ])
    if loss_function == 'LogCosh':
        classifier1.compile(optimizer = 'adam', 
                            loss = tf.keras.losses.LogCosh(), 
                            metrics = ['accuracy'])
    elif loss_function == 'binary_crossentropy':
        classifier1.compile(optimizer = 'adam', 
                            loss = 'binary_crossentropy', 
                            metrics = ['accuracy'])
    else:
        print('Incorrect input')
        return
    #classifier1.summary()
    classifier1.fit(x_train, y_train, 
                    batch_size = _batch_size, 
                    epochs = 100, 
                    verbose = 0)
    print(f'Accuracy on training set : {classifier1.evaluate(x_train, y_train)[1]}')
    print(f'Loss on training set : {classifier1.evaluate(x_train, y_train)[0]}')
    print()
    print(f'Accuracy on testing set : {classifier1.evaluate(x_test, y_test)[1]}')
    print(f'Loss on testing set : {classifier1.evaluate(x_test, y_test)[0]}')
    sample_case_performance(classifier1)
```
